```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe
df = pd.read_csv('data/processed_data.csv')

df = df.sample(n = 10, random_state=42)

df['cpvCode'] = df['cpvCode'].str.split('-', expand=True)[0]

# Select the features to use for clustering
X = df[['cpvCode', 'organizationId']].values

# Specify the number of clusters
k = 5

# Initialize the k-means algorithm with the number of clusters
kmeans = KMeans(n_clusters=k)

# Fit the algorithm to the data
kmeans.fit(X)

# Get the labels assigned to each data point
labels = kmeans.labels_

# Plot the clusters using a scatter plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Clusters')
plt.xlabel('CPV code')
plt.ylabel('Organization ID')
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe
df = pd.read_csv('data/processed_data.csv')

df['cpvCode'] = df['cpvCode'].str.split('-', expand=True)[0]

# Select the features to use for clustering
X = df[['cpvCode', 'organizationId']].values

# Compute the within-cluster sum of squares (WCSS) for different values of k
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values as a function of k
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe
df = pd.read_csv('data/processed_data.csv')

df['cpvCode'] = df['cpvCode'].str.split('-', expand=True)[0]

# One-hot encode the countyCode feature
df = pd.get_dummies(df, columns=['countyCode'])

# Select the features to use for clustering
X = df.drop(['countyCode'], axis=1).values

# Compute the within-cluster sum of squares (WCSS) for different values of k
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values as a function of k
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Specify the number of clusters
k = 3

# Initialize the k-means algorithm with the number of clusters
kmeans = KMeans(n_clusters=k)

# Fit the algorithm to the data
kmeans.fit(X)

# Get the labels assigned to each data point
labels = kmeans.labels_

# Print the labels
print(labels)

# Plot the clusters using a scatter plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Clusters')
plt.xlabel('CPV code')
plt.ylabel('County code')
plt.show()
```
