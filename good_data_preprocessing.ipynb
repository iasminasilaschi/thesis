```python
import pandas as pd

# Reading the data and saving it into the "data" variable

data = pd.read_csv(
    'data/sqllab_anunturi_de_atribuire_procedure_anunturi_de_participare_20220914T195330_v.1.0 - Copy.csv',
    sep=';')

data
```

```python
# Dropping all rows of data that contain NAN values for numeric columns

data = data.dropna()
data.reset_index(inplace=True, drop=True)

data
```

```python
from forex_python.converter import CurrencyRates
import datetime

c = CurrencyRates()

def get_exchange_rate(bid):
    date_str = bid['publicationDate']
    date = datetime.datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S').date()

    currency_code = bid['currencyCode']
    exchange_rate = c.get_rate(currency_code, 'RON', date)
    # print(f"Processing currency code '{currency_code}' on date '{date}'...")

    return exchange_rate

def exchange_values_to_RON(bid):
    currency_code = bid['currencyCode']
    if currency_code == 'RON':
        return bid
    exchange_rate = get_exchange_rate(bid)
    bid['contractValue'] = bid['contractValue'] * exchange_rate
    bid['cNoticeEstimatedContractValue'] = bid['cNoticeEstimatedContractValue'] * exchange_rate
    bid['currencyCode'] = 'RON'
    return bid

for i, bid in data.iterrows():
    data.loc[i] = exchange_values_to_RON(bid)

data.to_csv('data/processed_currency_data.csv', index=False)

data
```

```python
import pandas as pd

data = pd.read_csv('data/processed_currency_data.csv')

new_rows = []
rows_to_delete = []

for index, row in data.iterrows():
    if ',' in row['organizationName'] and ',' in row['organizationId']:
        organizationNames = row['organizationName'].split(',')
        organizationIds = row['organizationId'].split(',')
        if len(organizationNames) == len(organizationIds):
            for i in range(len(organizationNames)):
                new_row = row
                new_row['organizationName'] = organizationNames[i]
                new_row['organizationId'] = organizationIds[i]
                new_rows.append(new_row)
            rows_to_delete.append(index)
        else:
            rows_to_delete.append(index)

new_bids = data.append(new_rows, ignore_index=True)
bids_with_only_one_winner_per_row = new_bids.drop(rows_to_delete)

data = bids_with_only_one_winner_per_row

data
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('data/processed_currency_data.csv')

for index, row in data.iterrows():
    if ',' in row['organizationId']:
        data.drop(index)

data.drop_duplicates(inplace=True)
data.reset_index(inplace=True, drop=True)

data = data.drop(data[(data['contractValue'] == 0) | (data['cNoticeEstimatedContractValue'] == 0)].index)

data.to_csv('data/no_multiple_winners.csv', index=False)

# Split the cpvCode column
data['cpv'] = data['cpvCode'].str.split('-', expand=True)[0]
data['cpvCategory'] = data['cpv'].str.slice(stop=5)

data = data.drop([
    'cpv',
    'currencyCode',
    'cpvCodeType',
    'organizationName',
], axis=1)

# add one-hot encoded columns to the cpvCode categorical data
data_encoded = pd.get_dummies(data, columns=['cpvCode'])
data = data_encoded

# apply Label Encoding to the categorical columns
columns_to_encode = [
    'noticeNo',
    'title',
    'publicationDate',
    'contractingAuthorityName',
    'cpvCodeName',
    'cNoticeNo',
    'cNoticePublicationDate',
    'cNoticeTitle',
    'cpvCategory',
    'countyCode',
]

# add label encoding to the other categorical columns
label_encoder = LabelEncoder()
for col in columns_to_encode:
    data[col] = label_encoder.fit_transform(data[col])

data.to_csv('data/better_encoded_data_nmw.csv', index=False)

data
```
